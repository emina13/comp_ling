1. Напишите регулярное выражение, которое возвращает список первых двух букв каждого слова строки. Обратите внимание на работу с дефисом.

```python
result = re.findall(r'\b\w\w', text) #\b	граница слова, \w	любая цифра или буква. Слова из 1 буквы не показываются
result
```

2. Напишите регулярное выражение, которое выбирает из строки все слова, в которых строго больше 3 символов.

```python
re.findall(r'\S{4,}', text) #\S	любой непробельный символ, {n, m}	от n до m вхождений. Включаю все слова, у которых минимум 4 буквы.
```

3. Напишите регулярное выражение, которое заменит все подстроки, обозначающие
время (только время, не даты), в строке на TBD.


```python
data_text = 'Я полечу домой 27.06. Мой самолет взлетит в 01:20. Надеюсь, что ровно в 07:10:15 я уже приземлюсь'
pattern = r'((?:[01]\d|2[0-3])\:(?:[0-5]\d)(?:\:[0-5]\d)?)'  #() группирует выражение и возвращает найденный текст, 
#?	0 или 1 вхождение шаблона, стоящего перед ?, \d	любая цифра [0-9], a|b	равно a или b, [..]	один из символов в скобках
#\	экранирование спецсимволов
data_text = re.sub(pattern, "TBD", data_text) 

print(data_text)
```

4. Напишите регулярное выражение, которое заменяет произвольное количество
пробельных символов внутри строки на один пробел.

```python
probel = 'В                 1908                году был поставлен вопрос о                возможности введения в Туркестане общеимперской         административной системы управления.                   Рассматривалось предложение       о наделении генерал-губернатора края полномочиями         наместника. '
print(probel)
```

```python
pattern = r'\s{1,}' #\s	любой пробельный символ, {n, m}	от n до m вхождений
probel2 = re.sub(pattern, ' ', probel)

print(probel2)
```

5. Напишите регулярное выражение, которое удаляет идущие подряд повторы. Одно
слово из группы должно остаться.


```python
repeating = 'В в в в в в в в в 1908 году был поставлен вопрос о о о о о о о возможности введения в Туркестане Туркестане Туркестане общеимперской общеимперской общеимперской общеимперской административной системы управления. Рассматривалось предложение о наделении генерал-губернатора края полномочиями наместника наместника наместника наместника наместника. '
print(repeating)
```

```python
repeating2 = re.sub(r'\b([^\W\d_]+)(\s+\1)+\b', r'\1', repeating) #\b	граница слова, [^..]	любой символ, кроме тех, что в скобках
#\ экранирование спецсимволов, \d	любая цифра [0-9], \W	все, кроме букв и цифр
#\s	любой пробельный символ, \b	граница слова, +	1 и более вхождений шаблона, стоящего перед +
#\1	заменяет найденное совпадение единичным экземпляром найденного слова
print(repeating2)
```

поняла, что регистр имеет значение

7. Напишите регулярное выражение, которое возвращает список аббревиатур в строке

8. Напишите регулярное выражение, которое разделяет текст на предложения.


```python
sentence = 'В 1908 году был поставлен вопрос о возможности введения в Туркестане общеимперской административной системы управления. Рассматривалось предложение о наделении генерал-губернатора края полномочиями наместника. В 1911 г. началась разработка основных начал преобразования управления Туркестанским краем, однако реформирование не было произведено и власть в Туркестанском крае по-прежнему находилась под контролем Военного министерства. В начале 1913 года императором были одобрены разработанные Советом министров основные принципы преобразований в управлении Туркестанским краем, однако эти принципы так и не получило окончательного редакционного оформления и не прошли рассмотрение в Государственном Совете и Государственной Думе.'
re.split(r'(?<=\w[.?!]\s)', sentence) 
#\w	любая цифра или буква, \s	любой пробельный символ, [..]	один из символов в скобках, ?	0 или 1 вхождение шаблона, стоящего перед ?
```

интересно, что после точки как сокращения слова год как будто начинается новое предложение

9. Напишите регулярное выражение, которое определяет, что строка является номером
российского мобильного телефона любого оператора

10. Напишите регулярное выражение, которое проверяет, что все предложения в строке
начинаются с заглавной буквы.

**2 часть: Регулярные выражения для очистки корпуса текстов**
Изучите собранный вами в работе 1 корпус текстов и примените регулярные
выражения для очистки от лишней информации (тегов html-разметки, url, непонятных
символов и др.) или приведения отдельных потенциально важных элементов текстов к
единому виду (времени или дат, например).
Сохраните обновленный корпус в формате json.



```python
import json
```

```python
with open('tashkent.json') as f:
    data = json.load(f)
list(data.values())[0:10]
```

```python
corpus = str(list(data.values())[0])
for i in range(1, len(data)-1):
    corpus += ' '
    corpus += list(data.values())[i]
```

```python
corpus = re.sub(r'\n', ' ', corpus) #удалила переносы и проверила их отсутствие
corpus[:2500]
```

```python
corpus = re.sub(r'\([^()]*\)', '', corpus) #в скобках транскрипция слов, даты, происхождение слов и тд, они не нужны
corpus[:2500]
```

```python
corpus = re.sub(r'\s+(?=(?:[,.?!:;…]))', '', corpus)
corpus[:2500]
#в местах, где была удалена информация в скобках и сами скобки, мы видим лишние пробелы. устраняем
```

```python
corpus = re.sub(r'а́', 'а', corpus)  #часто название статьи пишется с ударением. избавляемся от него и проработаем букву ё
corpus = re.sub(r'А́', 'А', corpus)
corpus = re.sub(r'Ё', 'Е', corpus)
corpus = re.sub(r'ё', 'е', corpus)
corpus = re.sub(r'É', 'Е', corpus)
corpus = re.sub(r'е́', 'е', corpus)
corpus = re.sub(r'и́', 'и', corpus)
corpus = re.sub(r'И́', 'И', corpus)
corpus = re.sub(r'о́', 'о', corpus)
corpus = re.sub(r'О́', 'О', corpus)
corpus = re.sub(r'у́', 'у', corpus)
corpus = re.sub(r'У́', 'У', corpus)
corpus = re.sub(r'ы́', 'ы', corpus)
corpus = re.sub(r'Ы́', 'Ы', corpus)
corpus = re.sub(r'я́', 'я', corpus)
corpus = re.sub(r'Я́', 'Я', corpus)
corpus = re.sub(r'ю́', 'ю', corpus)
corpus = re.sub(r'Ю́', 'Ю', corpus)
corpus = re.sub(r'э́', 'э', corpus)
corpus = re.sub(r'Э́', 'Э', corpus)
```

```python
corpus = re.sub(r'\s{1,}', ' ', corpus) #на всякий случай, если между 2 саммари есть по несколько пробелов
corpus[:2500]
```

```python
corpus = re.sub(r'<(.*?)>', ' ', corpus) #убрала html разметку
```

```python
with open('corpus.json', 'w') as f:
    json.dump(corpus, f, ensure_ascii=False)
```
